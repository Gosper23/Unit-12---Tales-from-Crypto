## 1. Sentiment Analysis

Use the [newsapi](https://newsapi.org/) to pull the latest news articles for Bitcoin and Ethereum and create a DataFrame of sentiment scores for each coin.

Use descriptive statistics to answer the following questions:
1. Which coin had the highest mean positive score?
2. Which coin had the highest negative score?
3. Which coin had the highest positive score?

```python
# Initial imports
import os
import pandas as pd
from dotenv import load_dotenv
import nltk as nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

%matplotlib inline
```

```python
# Read your api key environment variable
load_dotenv()
api_key = os.getenv("NEWS_API_KEY")
```

```python
# Create a newsapi client
from newsapi import NewsApiClient 
newsapi = NewsApiClient(api_key=api_key)
```

```python
# Fetch the Bitcoin news articles
btc_headlines = newsapi.get_everything(q="bitcoin", language="en", sort_by="relevancy")
btc_headlines
```

```python
# Fetch the Ethereum news articles
eth_headlines = newsapi.get_everything(q="ethereum", language="en", sort_by="relevancy")
eth_headlines
```

```python
# Create the Bitcoin sentiment scores DataFrame

articles = []

for article in btc_headlines["articles"]:
    try:
        title = article["title"]
        text = article["content"]
        date = article["publishedAt"][:10]
        sentiment = analyzer.polarity_scores(text)
        compound = sentiment["compound"]
        pos = sentiment["pos"]
        neu = sentiment["neu"]
        neg = sentiment["neg"]
        
        articles.append({
            "title": title,
            "text": text,
            "date": date,
            "compound": compound,
            "positive": pos,
            "negative": neg,
            "neutral": neu
            
        })
        
    except AttributeError:
        pass
    
# Create DataFrame
btc_df = pd.DataFrame(articles)

# Reorder DataFrame columns
cols = ["date", "title", "text", "compound", "positive", "negative", "neutral"]
btc_df = btc_df[cols]

btc_df.head()
```

```python
# Create the Ethereum sentiment scores DataFrame

articles = []                                              # create variable to hold values

for article in eth_headlines["articles"]:                  # create loop to get data from articles
    try:
        title = article["title"]                           # Extract title
        text = article["content"]                          # Extract content
        date = article["publishedAt"][:10]                 # Extrade date of publishing
        sentiment = analyzer.polarity_scores(text)         # Use analyser to create polarity
        compound = sentiment["compound"]                   # Extract compound score from polarity scores
        pos = sentiment["pos"]                             # Extract positive score from polarity scores
        neu = sentiment["neu"]                             # Extract neutral score from polarity scores
        neg = sentiment["neg"]                             # Extract negative score from polarity scores
        
        articles.append({
            "title": title,
            "text": text,
            "date": date,
            "compound": compound,
            "positive": pos,
            "negative": neg,
            "neutral": neu
            
        })
        
    except AttributeError:
        pass
    
# Create DataFrame
eth_df = pd.DataFrame(articles)

# Reorder DataFrame columns
cols = ["date", "title", "text", "compound", "positive", "negative", "neutral"]
eth_df = eth_df[cols]

eth_df.head()
```

```python
# Describe the Bitcoin Sentiment
btc_df.describe()
```

```python
# Describe the Ethereum Sentiment
eth_df.describe()
```

## 2. Natural Language Processing
---
###   Tokenizer

In this section, you will use NLTK and Python to tokenize the text for each coin. Be sure to:
1. Lowercase each word.
2. Remove Punctuation.
3. Remove Stopwords.

```python
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
from string import punctuation
import re
```

```python
# Instantiate the lemmatizer
lemmatizer = WordNetLemmatizer()

# Create a list of stopwords
sw = set(stopwords.words('english'))
sw
# Expand the default stopwords list if necessary
sw_addons = {'pas', 'char', 'z', 'ra', 'h', 'wa', 'cc', 'sa', 'gmt', 'w', 'air', 'mex', 'ha' , 'q', 'eked'}
```

```python
# Complete the tokenizer function
def tokenizer(text):
    """Tokenizes text."""
    
    # Remove the punctuation from text
    regex = re.compile("[^a-zA-Z]")
    re_clean = regex.sub(' ', text)
   
    # Create a tokenized list of the words
    words = word_tokenize(re_clean)
    
    # Lemmatize words into root words
    lem = [lemmatizer.lemmatize(word) for word in words]
   
    # Convert the words to lowercase
    word_lower = [word.lower() for word in lem]
    
    # Remove the stop words
    output1 = [word.lower() for word in word_lower if word.lower() not in sw]
    output2 = [word.lower() for word in output1 if word.lower() not in sw_addons]
    return output2
```

```python
# Create a new tokens column for Bitcoin
btc_tokens = btc_df.copy()                                  # Create variable
tokenized = []                                              # Variable to hold tokenized values

for article in btc_df['text']:                              # Loop through articles 
    tokenized_articles = tokenizer(article)                 # tokenise article using tokenizer
    tokenized.append({'tokens':tokenized_articles})         # append all tokens to list of dics

btc_tokens['tokens'] = pd.DataFrame(tokenized)              # combin tokens with dataframe
btc_tokens.head()
```

```python
# Create a new tokens column for Ethereum
eth_tokens = eth_df.copy()                                    # Create variable
tokenized = []                                                # Variable to hold tokenized values

for article in eth_df['text']:                                # Loop through articles 
    tokenized_articles = tokenizer(article)                   # tokenise article using tokenizer
    tokenized.append({'tokens':tokenized_articles})           # append all tokens to list of dics

eth_tokens['tokens'] = pd.DataFrame(tokenized)                # combin tokens with dataframe
eth_tokens.head()
```

### NGrams and Frequency Analysis

In this section you will look at the ngrams and word frequency for each coin. 

1. Use NLTK to produce the n-grams for N = 2. 
2. List the top 10 words for each coin. 

```python
from collections import Counter
from nltk import ngrams
```

```python
# Generate the Bitcoin N-grams where N=2
btc_bigrams = [Counter(ngrams(tokens, n=2)) for tokens in btc_tokens['tokens']]
btc_bigrams[1]
```

```python
# Generate the Ethereum N-grams where N=2
eth_bigrams = [Counter(ngrams(tokens, n=2)) for tokens in eth_tokens['tokens']]
eth_bigrams[1]
```

```python
# Function token_count generates the top 10 words for a given coin
def token_count(tokens, N=3):
    """Returns the top N tokens from the frequency count"""
    return Counter(tokens).most_common(N)
```

```python
# Use token_count to get the top 10 words for Bitcoin
all_btc_tokens = []                                   # List vairables for tokens

for tokens in btc_tokens['tokens']:                   # Loop through articles in column 
    all_btc_tokens += tokens                          # add to variable all_eth_tokens

top10_btc_words = token_count(all_btc_tokens, 10)     # Get top 10 words through token_count
top10_btc_words
```

```python
# Use token_count to get the top 10 words for Ethereum
all_eth_tokens = []                                   # List vairables for tokens

for tokens in eth_tokens['tokens']:                   # Loop through articles in column 
    all_eth_tokens += tokens                          # add to variable all_eth_tokens

top10_eth_words = token_count(all_eth_tokens, 10)     # Get top 10 words through token_count
top10_eth_words
```

## 3. Named Entity Recognition

In this section, you will build a named entity recognition model for both Bitcoin and Ethereum, then visualize the tags using SpaCy.

```python
import spacy
from spacy import displacy
```

```python
# Download the language model for SpaCy
# !python -m spacy download en_core_web_sm
```

```python
# Load the spaCy model
nlp = spacy.load('en_core_web_sm')
```

---
### Bitcoin NER

```python
# Concatenate all of the Bitcoin text together
bitcoin_text = []                                           # Create Variable to hold text value

for text in btc_tokens['text']:                             # Create loop to go through eth_tokens in column
    bitcoin_text.append(text)                               # Append the text to variable
    
bitcoin_text_string = ' '.join(bitcoin_text)                # convert list to one long string
```

```python
# Run the NER processor on all of the text
bitcoin_ner = nlp(bitcoin_text_string)

# Add a title to the document
bitcoin_ner.user_data['title'] = "Bitcoin NER"
```

```python
# Render the visualization
displacy.render(bitcoin_ner, style='ent')
```

```python
# List all Entities
bitcoin_ents = set([ent.text for ent in bitcoin_ner.ents])
print(bitcoin_ents)
```
